ðŸ“‚ High-Level Architecture Summary
| Layer          | Folder                | Purpose              | What It Does                          | Why It Exists                    |
| -------------- | --------------------- | -------------------- | ------------------------------------- | -------------------------------- |
| Data Source    | `data/raw/`           | Source of truth      | Stores original CSV files             | Prevents modifying original data |
| Processed Data | `data/processed/`     | Cleaned outputs      | Stores engineered features, snapshots | Used for ML & reporting          |
| Ingestion      | `src/ingestion/`      | Data loading         | Reads raw CSVs                        | Standardizes input data          |
| Transformation | `src/transformation/` | Relational modeling  | Builds object relationships           | Creates structured data graph    |
| Health         | `src/health/`         | Health scoring logic | Calculates device/interface health    | Business logic layer             |
| Features       | `src/features/`       | ML features          | Generates ML-ready features           | Converts raw data â†’ model input  |
| Models         | `src/models/`         | Machine learning     | Trains & evaluates models             | Intelligence layer               |
| Inference      | `src/inference/`      | Predictions          | Runs trained models                   | Real-time scoring                |
| UI             | `app/`                | Streamlit dashboard  | Visualizes system health              | User interaction layer           |
| Orchestrator   | `main.py`             | Pipeline runner      | Runs full workflow                    | Entry point                      |


```python
# -------------------------------------------------------
# Exploratory Data Analysis (EDA)
# Intelligent Device Health Monitoring System
# Cleaned Version: Top 10 Countries & Correct Stored vs Actual Counts
# -------------------------------------------------------

# -----------------------
# 1. Imports & Setup
# -----------------------
import sys
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (12,6)

REPO_ROOT = Path("../")
SRC_PATH = REPO_ROOT / "src"
sys.path.append(str(SRC_PATH))
from transformation.load_relational_data import load_all_data

# -----------------------
# 2. Load Data
# -----------------------
db = load_all_data()

organizations = db['organizations']
assets = db['assets']
devices = db['devices']
device_classes = db['device_classes']
interfaces = db['interfaces']
events = db['events']

# -----------------------
# 3. Convert to DataFrames with Country
# -----------------------
# Organizations
org_df = pd.DataFrame([{
    "organization_id": o.organization_id,
    "org_name": o.org_name,
    "org_industry": o.org_industry,
    "org_country": getattr(o, "org_country", ""),
    "num_assets": len(o.assets)
} for o in organizations.values()])

# Top 10 countries
top_countries = org_df['org_country'].value_counts().nlargest(10).index
org_df_top = org_df[org_df['org_country'].isin(top_countries)]

# Assets
asset_df = pd.DataFrame([{
    "asset_id": a.asset_id,
    "asset_name": a.asset_name,
    "organization_id": a.organization.organization_id if a.organization else None,
    "org_country": a.organization.org_country if a.organization else None,
    "asset_location": a.asset_location,
    "asset_owner": a.asset_owner,
    "asset_purchase_date": a.asset_purchase_date,
    "num_devices": len(a.devices)
} for a in assets.values()])
asset_df["asset_purchase_date"] = pd.to_datetime(asset_df["asset_purchase_date"])
asset_df_top = asset_df[asset_df['org_country'].isin(top_countries)]

# Devices
device_df = pd.DataFrame([{
    "device_id": d.device_id,
    "device_ip": d.device_ip,
    "device_serial": d.device_serial,
    "device_manufacturer": d.device_manufacturer,
    "asset_id": d.asset.asset_id if d.asset else None,
    "org_country": d.asset.organization.org_country if d.asset and d.asset.organization else None,
    "device_class_id": d.device_class.device_class_id if d.device_class else None,
    "num_interfaces": len(d.interfaces),
    "num_events": len(d.events)
} for d in devices.values()])
device_df_top = device_df[device_df['org_country'].isin(top_countries)]

# Device Classes
device_class_df = pd.DataFrame([{
    "device_class_id": dc.device_class_id,
    "device_class_name": dc.device_class_name,
    "num_devices": len(dc.devices)
} for dc in device_classes.values()])

# Interfaces
interface_df = pd.DataFrame([{
    "interface_id": i.interface_id,
    "interface_name": i.interface_name,
    "device_id": i.device.device_id if i.device else None,
    "org_country": i.device.asset.organization.org_country if i.device and i.device.asset and i.device.asset.organization else None,
    "interface_status": i.interface_status,
    "num_events": len(i.events)
} for i in interfaces.values()])
interface_df_top = interface_df[interface_df['org_country'].isin(top_countries)]

# Events (subset)
event_sample_size = 5000
event_df_full = pd.DataFrame([{
    "event_id": e.event_id,
    "event_timestamp": pd.to_datetime(e.event_timestamp),
    "device_id": e.device.device_id if e.device else None,
    "interface_id": e.interface.interface_id if e.interface else None,
    "org_country": e.device.asset.organization.org_country if e.device and e.device.asset and e.device.asset.organization else None,
    "event_type": e.event_type,
    "event_description": e.event_description
} for e in events.values()])
event_df_top = event_df_full[event_df_full['org_country'].isin(top_countries)]
event_df = event_df_top.sample(n=min(event_sample_size, len(event_df_top)), random_state=42)

# -----------------------
# 4. EDA - Organizations
# -----------------------
sns.countplot(data=org_df_top, x="org_industry", hue="org_country")
plt.title("Organization Industry Distribution (Top 10 Countries)")
plt.xticks(rotation=90)
plt.legend(title="Country", bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()

sns.histplot(org_df_top["num_assets"], bins=20)
plt.title("Number of Assets per Organization (Top 10 Countries)")
plt.tight_layout()
plt.show()

sns.countplot(data=org_df_top, x="org_country")
plt.title("Organization Count by Country (Top 10)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# -----------------------
# 5. EDA - Assets
# -----------------------
sns.histplot(asset_df_top["num_devices"], bins=20)
plt.title("Number of Devices per Asset (Top 10 Countries)")
plt.tight_layout()
plt.show()

sns.histplot(asset_df_top["asset_purchase_date"].dropna(), bins=20, kde=True)
plt.title("Asset Purchase Date Distribution (Top 10 Countries)")
plt.tight_layout()
plt.show()

sns.countplot(data=asset_df_top, x="org_country")
plt.title("Assets Count by Country (Top 10)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# -----------------------
# 6. EDA - Device Classes & Devices
# -----------------------
sns.barplot(data=device_class_df, x="device_class_name", y="num_devices")
plt.xticks(rotation=90)
plt.title("Number of Devices per Device Class")
plt.tight_layout()
plt.show()

sns.histplot(device_df_top["num_interfaces"], bins=20)
plt.title("Number of Interfaces per Device (Top 10 Countries)")
plt.tight_layout()
plt.show()

sns.histplot(device_df_top["num_events"], bins=20)
plt.title("Number of Events per Device (Top 10 Countries)")
plt.tight_layout()
plt.show()

sns.countplot(data=device_df_top, x="org_country")
plt.title("Devices Count by Country (Top 10)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# -----------------------
# 7. EDA - Interfaces
# -----------------------
sns.countplot(data=interface_df_top, x="interface_status", hue="org_country")
plt.title("Interface Status Distribution by Country (Top 10)")
plt.xticks(rotation=90)
plt.legend(title="Country", bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()

sns.histplot(interface_df_top["num_events"], bins=20)
plt.title("Number of Events per Interface (Top 10 Countries)")
plt.tight_layout()
plt.show()

# -----------------------
# 8. EDA - Events
# -----------------------
sns.histplot(event_df["event_timestamp"], bins=50)
plt.title("Events Over Time (sampled, Top 10 Countries)")
plt.tight_layout()
plt.show()

sns.countplot(data=event_df, x="event_type", hue="org_country")
plt.title("Event Type Distribution by Country (Top 10, Sampled)")
plt.xticks(rotation=90)
plt.legend(title="Country", bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()

# -----------------------
# 9. Relationships - Aggregates with Country
# -----------------------

# ---- Assets â†’ Devices ----
asset_device_counts = interface_df_top.groupby("device_id")["interface_id"].count().reset_index()  # intermediate
device_interface_counts = interface_df_top.groupby("device_id")["interface_id"].count().reset_index()  # used below
# Correct merged_asset using actual device counts
asset_device_counts = device_df_top.groupby(["asset_id","org_country"])["device_id"].count().reset_index().rename(columns={"device_id":"device_count_actual"})
merged_asset = asset_df_top.merge(asset_device_counts, on=["asset_id","org_country"], how="left")
merged_asset["device_count_actual"].fillna(0, inplace=True)

sns.scatterplot(data=merged_asset, x="num_devices", y="device_count_actual", hue="org_country")
plt.title("Asset: Stored vs Actual Device Count (Top 10 Countries)")
plt.tight_layout()
plt.legend(title="Country", bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()

# ---- Devices â†’ Interfaces ----
# Recompute stored num_interfaces using actual interface_df_top
stored_interfaces = interface_df_top.groupby("device_id")["interface_id"].count().reset_index()
stored_interfaces.rename(columns={"interface_id": "num_interfaces"}, inplace=True)

# Merge with device_df_top
merged_device = device_df_top.drop(columns=["num_interfaces"]).merge(stored_interfaces, on="device_id", how="left")
merged_device["num_interfaces"].fillna(0, inplace=True)

# Actual interface count (same as stored here)
device_interface_counts = interface_df_top.groupby(["device_id","org_country"])["interface_id"].count().reset_index().rename(columns={"interface_id":"interface_count_actual"})
merged_device = merged_device.merge(device_interface_counts, on=["device_id","org_country"], how="left")
merged_device["interface_count_actual"].fillna(0, inplace=True)

sns.scatterplot(data=merged_device, x="num_interfaces", y="interface_count_actual", hue="org_country")
plt.title("Device: Stored vs Actual Interface Count (Top 10 Countries)")
plt.tight_layout()
plt.legend(title="Country", bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()

# ---- Devices â†’ Events ----
# Stored num_events from device_df_top
stored_events = device_df_top.groupby("device_id")["num_events"].sum().reset_index()
stored_events.rename(columns={"num_events":"num_events_stored"}, inplace=True)

# Actual event count from event_df_top
device_event_counts = event_df_top.groupby(["device_id","org_country"])["event_id"].count().reset_index().rename(columns={"event_id":"event_count_actual"})

merged_device_event = device_df_top.drop(columns=["num_events"]).merge(stored_events, on="device_id", how="left")
merged_device_event = merged_device_event.merge(device_event_counts, on=["device_id","org_country"], how="left")
merged_device_event["event_count_actual"].fillna(0, inplace=True)

sns.scatterplot(data=merged_device_event, x="num_events_stored", y="event_count_actual", hue="org_country")
plt.title("Device: Stored vs Actual Event Count (Top 10 Countries)")
plt.tight_layout()
plt.legend(title="Country", bbox_to_anchor=(1.05,1), loc='upper left')
plt.show()

# -----------------------
# 10. Save Snapshots
# -----------------------
processed_dir = REPO_ROOT / "data" / "processed"
asset_df.to_csv(processed_dir / "assets_snapshot.csv", index=False)
device_df.to_csv(processed_dir / "devices_snapshot.csv", index=False)
event_df.to_csv(processed_dir / "events_snapshot_sample.csv", index=False)
```
